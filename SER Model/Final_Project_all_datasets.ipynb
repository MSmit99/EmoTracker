## Process Ravdess Dataset


```python
import os
import pandas as pd

# 1. Initialize an empty list
ravdess_data = []

# 2. Define ravdess_dataset_path
ravdess_dataset_path = 'ravdess-emotional-speech-audio'

# 3. Create dictionaries for mapping
emotion_mapping_ravdess = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fearful',
    '07': 'disgust',
    '08': 'surprised'
}

intensity_mapping_ravdess = {
    '01': 'normal',
    '02': 'strong'
}

# 4. Walk through the directory
for root, dirs, files in os.walk(ravdess_dataset_path):
    for file in files:
        if file.endswith(".wav"):
            # a. Construct the full file_path
            file_path = os.path.join(root, file)

            # b. Parse the filename
            # Filename format example: 03-01-01-01-01-01-01.wav
            # Modality-VocalChannel-Emotion-EmotionalIntensity-Statement-Repetition-Actor
            parts = file.replace(".wav", "").split("-")

            emotion_code = parts[2]
            intensity_code = parts[3]
            actor_id_str = parts[6]

            # c. Use mapping dictionaries
            emotion = emotion_mapping_ravdess.get(emotion_code, 'unknown')
            intensity = intensity_mapping_ravdess.get(intensity_code, 'unknown')

            # d. Determine actor ID and gender
            actor_id = int(actor_id_str)
            gender = 'male' if actor_id % 2 == 1 else 'female'

            # e. Append to ravdess_data
            ravdess_data.append([
                file_path,
                actor_id,
                emotion,
                intensity,
                gender,
                'Ravdess' # Dataset name
            ])

# 6. Create a pandas DataFrame
ravdess_df = pd.DataFrame(ravdess_data, columns=[
    'filepath',
    'actor_id',
    'emotion',
    'intensity',
    'gender',
    'dataset'
])

# 7. Convert the data types
ravdess_df['filepath'] = ravdess_df['filepath'].astype('string')
ravdess_df['actor_id'] = ravdess_df['actor_id'].astype('int32')
ravdess_df['emotion'] = ravdess_df['emotion'].astype('string')
ravdess_df['intensity'] = ravdess_df['intensity'].astype('string')
ravdess_df['gender'] = ravdess_df['gender'].astype('string')
ravdess_df['dataset'] = ravdess_df['dataset'].astype('string')

# 8. Print the head and info of the ravdess_df DataFrame
print("Ravdess DataFrame Head:")
print(ravdess_df.head())
print("\nRavdess DataFrame Info:")
ravdess_df.info()
```

## Process SAVEE Dataset

```python
import os
import pandas as pd
import re

# 1. Initialize an empty list
savee_data = []

# 2. Define the path to the SAVEE dataset
savee_dataset_path = 'surrey-audiovisual-expressed-emotion-savee/ALL'

# 3. Create dictionaries for mapping
emotion_mapping_savee = {
    'a': 'anger',
    'd': 'disgust',
    'f': 'fear',
    'h': 'happy',
    'n': 'neutral',
    'sa': 'sad',
    'su': 'surprise'
}

# 4. Walk through the directory
for root, dirs, files in os.walk(savee_dataset_path):
    for file in files:
        if file.endswith(".wav"):
            # a. Construct the full file_path
            file_path = os.path.join(root, file)

            # b. Extract the filename without the '.wav' extension.
            filename_without_ext = file.replace(".wav", "")

            # c. Parse the filename to get the actor ID, emotion code, and statement number
            # Filename format example: ID_EmotionNN.wav (e.g., 'DC_a01.wav', 'JE_sa15.wav')
            parts = filename_without_ext.split('_')

            actor_id = parts[0]
            emotion_statement_code = parts[1]

            # Extract emotion code (e.g., 'a', 'sa')
            emotion_code_match = re.match(r'([a-z]+)', emotion_statement_code)
            if emotion_code_match:
                emotion_code_short = emotion_code_match.group(1)
            else:
                emotion_code_short = 'unknown'

            # Extract statement number (e.g., '01', '15')
            statement_match = re.search(r'(\d+)', emotion_statement_code)
            if statement_match:
                statement_number = statement_match.group(1)
            else:
                statement_number = 'unknown'

            # d. Use mapping dictionaries for emotion
            emotion = emotion_mapping_savee.get(emotion_code_short, 'unknown')

            # e. Append to savee_data
            savee_data.append([
                file_path,
                actor_id,
                emotion,
                statement_number,
                'male', # all genders are male in the SAVEE dataset
                'SAVEE' # Dataset name
            ])

# 5. Create a pandas DataFrame
savee_df = pd.DataFrame(savee_data, columns=[
    'filepath',
    'actor_id',
    'emotion',
    'statement',
    'gender',
    'dataset'
])

# 6. Convert the data types
savee_df['filepath'] = savee_df['filepath'].astype('string')
savee_df['actor_id'] = savee_df['actor_id'].astype('string')
savee_df['emotion'] = savee_df['emotion'].astype('string')
savee_df['statement'] = savee_df['statement'].astype('string')
savee_df['dataset'] = savee_df['dataset'].astype('string')

# 7. Print the head and info of the savee_df DataFrame
print("SAVEE DataFrame Head:")
print(savee_df.head())
print("\nSAVEE DataFrame Info:")
savee_df.info()
```

## Process TESS Dataset

```python
import os
import pandas as pd
import re

# 1. Initialize an empty list
tess_data = []

# 2. Define the path to the TESS dataset
tess_dataset_path = 'toronto-emotional-speech-set-tess/TESS Toronto emotional speech set data'

# 3. Iterate through the dataset directory
for root, dirs, files in os.walk(tess_dataset_path):
    for file in files:
        if file.endswith(".wav"):
            # a. Construct the full file_path
            file_path = os.path.join(root, file)

            # b. Extract the emotion from the parent directory name
            # Example: 'OAF_angry' -> 'angry'
            emotion_dir = os.path.basename(root)
            emotion_parts = emotion_dir.split('_')
            emotion = emotion_parts[-1] # Last part is the emotion

            # c. Extract the filename without the '.wav' extension.
            filename_without_ext = file.replace(".wav", "")

            # d. Parse the filename to get the speaker ID and spoken word
            # Filename format example: Speaker_Word_Emotion.wav (e.g., 'OAF_back_angry.wav')
            filename_parts = filename_without_ext.split('_')

            speaker_id = filename_parts[0] # e.g., 'OAF', 'YAF'
            # The word can be composed of multiple parts before the emotion
            word = '_'.join(filename_parts[1:-1]) # All parts between speaker_id and emotion

            # e. Append to tess_data
            tess_data.append([
                file_path,
                speaker_id,
                word,
                emotion,
                'female', # all genders are female in the TESS dataset
                'TESS' # Dataset name
            ])

# 5. Create a pandas DataFrame
tess_df = pd.DataFrame(tess_data, columns=[
    'filepath',
    'speaker_id',
    'word',
    'emotion',
    'gender',
    'dataset'
])

# 6. Convert the data types
tess_df['filepath'] = tess_df['filepath'].astype('string')
tess_df['speaker_id'] = tess_df['speaker_id'].astype('string')
tess_df['word'] = tess_df['word'].astype('string')
tess_df['emotion'] = tess_df['emotion'].astype('string')
tess_df['dataset'] = tess_df['dataset'].astype('string')

# 7. Print the head of the tess_df DataFrame
print("TESS DataFrame Head:")
print(tess_df.head())
print("\n")
# 8. Print the information about the tess_df DataFrame
print("TESS DataFrame Info:")
tess_df.info()
```

```python
import librosa
import numpy as np
import pandas as pd

def extract_features(file_path, max_pad_len=174, n_mfcc=40):
    """
    Extracts MFCC features from an audio file.

    Returns a 2D array of shape (n_mfcc, max_pad_len) —
    this becomes one "time-series" sample for the LSTM.
    """
    try:
        audio, sample_rate = librosa.load(file_path, res_type='kaiser_fast')

        # Extract MFCCs — shape: (n_mfcc, time_steps)
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=n_mfcc)

        # Pad or truncate to a fixed length so all samples are the same shape
        pad_width = max_pad_len - mfccs.shape[1]
        if pad_width > 0:
            mfccs = np.pad(mfccs, pad_width=((0, 0), (0, pad_width)), mode='constant')
        else:
            mfccs = mfccs[:, :max_pad_len]

        return mfccs  # shape: (40, 174)

    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None
```

```python
# Apply feature extraction to every audio file
cleaned_df['features'] = cleaned_df['filepath'].apply(extract_features)

# Drop any files that failed to load
cleaned_df = cleaned_df.dropna(subset=['features'])

# Stack into a numpy array — shape: (n_samples, n_mfcc, time_steps)
X = np.array(cleaned_df['features'].tolist())  # e.g., (1440, 40, 174)

print(f"Feature matrix shape: {X.shape}")
# Output: (n_samples, 40, 174)
```

```python
def extract_mfcc_summary(mfcc):
    return {
        'mfcc_mean': np.mean(mfcc),
        'mfcc_std': np.std(mfcc),
        'mfcc_max': np.max(mfcc),
        'mfcc_min': np.min(mfcc)
    }

mfcc_summary = cleaned_df['features'].apply(extract_mfcc_summary)
mfcc_df = pd.DataFrame(mfcc_summary.tolist())

cleaned_df = pd.concat([cleaned_df, mfcc_df], axis=1)
```

```python
import tensorflow as tf

def create_tf_dataset(X, y, batch_size=32, training=False):
    """
    Converts numpy arrays into an optimized tf.data.Dataset.

    Args:
        X: Feature array of shape (n_samples, timesteps, features)
        y: One-hot label array of shape (n_samples, num_classes)
        batch_size: Number of samples per batch
        training: If True, shuffles the data each epoch
    """
    dataset = tf.data.Dataset.from_tensor_slices((X, y))

    if training:
        # Shuffle with a buffer = full dataset size for true randomness
        dataset = dataset.shuffle(buffer_size=len(X), reshuffle_each_iteration=True)

    # Batch before caching so cached units are already batched
    dataset = dataset.batch(batch_size)

    # .cache() — stores batches in RAM after first epoch
    # Pass a file path string to cache to disk instead, e.g. cache('/tmp/ser_cache')
    # Use disk caching if your dataset is too large to fit in RAM
    dataset = dataset.cache()

    # .prefetch() — AUTOTUNE lets TensorFlow dynamically decide
    # how many batches to prepare ahead based on GPU speed
    dataset = dataset.prefetch(buffer_size=tf.data.AUTOTUNE)

    return dataset
```

### Train and test BiLSTM-DNN model

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Bidirectional, LSTM, Dense, Dropout,
    BatchNormalization, GlobalAveragePooling1D, Add, Activation
)
from tensorflow.keras.regularizers import l2

def build_bilstm_dnn(input_shape, num_classes, dropout_rate=0.4):
    """
    Architecture:
      Input → BiLSTM stack (sequence modeling)
             → Global pooling (collapse time axis)
             → Deep DNN with residual connections
             → Softmax output
    """
    inputs = Input(shape=input_shape)  # (174, 40)

    # ── BiLSTM Block 1 ──────────────────────────────────────────
    x = Bidirectional(
        LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.2)
    )(inputs)
    x = BatchNormalization()(x)

    # ── BiLSTM Block 2 ──────────────────────────────────────────
    x = Bidirectional(
        LSTM(64, return_sequences=True, dropout=0.3, recurrent_dropout=0.2)
    )(x)
    x = BatchNormalization()(x)

    # ── BiLSTM Block 3 (final — no return_sequences) ────────────
    x = Bidirectional(
        LSTM(32, return_sequences=True, dropout=0.2)
    )(x)
    x = BatchNormalization()(x)

    # Collapse time dimension
    x = GlobalAveragePooling1D()(x)

    # ── Deep DNN Block ───────────────────────────────────────────
    # Layer 1
    x = Dense(256, kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(dropout_rate)(x)

    # Layer 2 with residual shortcut
    shortcut = x
    x = Dense(256, kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(dropout_rate)(x)
    x = Dense(256)(x)
    x = Add()([x, shortcut])  # residual connection
    x = Activation('relu')(x)

    # Layer 3
    x = Dense(128, kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(dropout_rate)(x)

    # Layer 4
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.2)(x)

    # ── Output ───────────────────────────────────────────────────
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model


input_shape = (X_train_scaled.shape[1], X_train_scaled.shape[2])  # (174, 40)
model = build_bilstm_dnn(input_shape, num_classes)
model.summary()
```

### Train and evaluate DNN-BiLSTM model

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Bidirectional, LSTM, Dense, Dropout,
    BatchNormalization, GlobalAveragePooling1D, Add, Activation, TimeDistributed
)
from tensorflow.keras.regularizers import l2

def build_dnn_bilstm(input_shape, num_classes, dropout_rate=0.4):
    """
    Architecture:
      Input → TimeDistributed DNN stack (feature transformation per timestep)
             → BiLSTM stack (sequence modeling)
             → Global pooling (collapse time axis)
             → Deep DNN (classification)
             → Softmax output
    """
    inputs = Input(shape=input_shape)  # (174, 40)

    # ── TimeDistributed DNN Block (Feature Transformation) ───────────────────────────
    # Layer 1
    x = TimeDistributed(Dense(64, kernel_regularizer=l2(1e-4)))(inputs)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(Activation('relu'))(x)
    x = TimeDistributed(Dropout(dropout_rate))(x)

    # Layer 2 with residual shortcut
    shortcut = x
    x = TimeDistributed(Dense(64, kernel_regularizer=l2(1e-4)))(x)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(Activation('relu'))(x)
    x = TimeDistributed(Dropout(dropout_rate))(x)
    x = TimeDistributed(Dense(64))(x)
    x = Add()([x, shortcut])  # residual connection
    x = TimeDistributed(Activation('relu'))(x)

    # Layer 3
    x = TimeDistributed(Dense(32, kernel_regularizer=l2(1e-4)))(x)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(Activation('relu'))(x)
    x = TimeDistributed(Dropout(dropout_rate))(x)

    # ── BiLSTM Block 1 ──────────────────────────────────────────
    x = Bidirectional(
        LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.2)
    )(x)
    x = BatchNormalization()(x)

    # ── BiLSTM Block 2 ──────────────────────────────────────────
    x = Bidirectional(
        LSTM(64, return_sequences=True, dropout=0.3, recurrent_dropout=0.2)
    )(x)
    x = BatchNormalization()(x)

    # ── BiLSTM Block 3 (final — no return_sequences) ────────────
    x = Bidirectional(
        LSTM(32, return_sequences=True, dropout=0.2)
    )(x)
    x = BatchNormalization()(x)

    # Collapse time dimension
    x = GlobalAveragePooling1D()(x)

    # ── Final DNN Block (Classification) ───────────────────────────
    # Layer 1
    x = Dense(256, kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(dropout_rate)(x)

    # Layer 2 with residual shortcut
    shortcut = x
    x = Dense(256, kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(dropout_rate)(x)
    x = Dense(256)(x)
    x = Add()([x, shortcut])  # residual connection
    x = Activation('relu')(x)

    # Layer 3
    x = Dense(128, kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(dropout_rate)(x)

    # Layer 4
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.2)(x)

    # ── Output ───────────────────────────────────────────────────
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model


# Example usage with input_shape and num_classes
input_shape_for_dnn_bilstm = (X_train_scaled.shape[1], X_train_scaled.shape[2])  # (174, 40)
model_dnn_bilstm = build_dnn_bilstm(input_shape_for_dnn_bilstm, num_classes)
model_dnn_bilstm.summary()
```

### Create pipelined function to pass any audio clip file for analysis

```python
def predict_emotion(file_path, model, scaler, label_encoder, max_pad_len=174, n_mfcc=40):
    features = extract_features(file_path, max_pad_len, n_mfcc)
    if features is None:
        return None

    # Transpose: (n_mfcc, time) → (time, n_mfcc)
    features = features.T  # (174, 40)

    # Normalize using the fitted scaler
    features_scaled = scaler.transform(features)

    # Add batch dimension → (1, 174, 40)
    features_input = np.expand_dims(features_scaled, axis=0)

    prediction = model.predict(features_input)
    emotion_idx = np.argmax(prediction)
    emotion = label_encoder.inverse_transform([emotion_idx])[0]
    confidence = prediction[0][emotion_idx]

    return emotion, confidence, prediction


```
